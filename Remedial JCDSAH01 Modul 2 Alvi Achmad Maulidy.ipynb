```
m = df.isnull().sum()
m[0:13]
#menunjukan tidak ada missing data point
```

```
#membuat kategori baru
df1 = []
for item in df['alcohol']:
    if item > 0 and item <= 10:
        df1.append('mild')
    elif item > 10 and item < 11:
        df1.append('moderate')
    elif item >= 11:
        df1.append('strong')
```

```
df['alcohol_level'] = df1
```

```
df.head()
```

```
plt.figure(figsize=(10,10))
sns.set_style('darkgrid')
sns.lmplot(x = 'fixed_acidity', y = 'pH', data=df, hue = 'alcohol_level', palette = 'coolwarm', markers = ['o', 'v','*'])
plt.title('Coleration between Acidity and pH level in Wine')

#Analisis: dalam plot tersebut kita bisa lihat hubungan antara keasaman dan nilai pH. Semakin tinggi nilai keasaman wine semakin rendah nilai pHnya, dikarenakan keasaman suatu susbtan ditentukan oleh tinggi atau rendahnya nilai pH, namun itu tidak berpengaruh terhadap level Alkohol pada wine.
```

```
sns.scatterplot(hue = 'alcohol_level', y= 'fixed_acidity', x = 'density', data = df)
plt.title('Acidity and Density in Wine')

#Analisis: Massa jenis pada Wine memeliki ketergantungan terhadap nilai keasamaan pada Wine. Chart tersebut juga menunjukan distribusi kualitas pada Wine, wine yang memiliki massa jenis dan nilai keasamaan yang lebih rendah cenderung memiliki tingkatan alkohol yang kuat.
```

```
sns.jointplot(x = 'residual_sugar', y = 'pH', data = df, kind = 'hex')
plt.title('pH and Residual Sugar in Wine')

#Analisis: chart tersebut menunjukan distribusi wine dengan hubungan tingkatan residual sugar dan nilai pH.
```

```
sns.lmplot(x = 'residual_sugar', y = 'volatile_acidity', col = 'alcohol_level', row = 'quality_cat', data = df)
#analisis: chart-chart tersebut menunjukan distribusi macam-macam wine, yang berdasarkan tingkatan violatile acidity dan residual sugar atau sisa-sisa gula yang terdapat dalam wine, dikategorikan melalui kualitas dan tingkat alkohol pada wine. Dalam chart-chart tersebut menampilkan pendistribusian data yang merata berdasarkan residual sugar, tetapi menunjukan pendistribusian yang bervariasi berdasarkan tingkatan volatile acidity pada wine.
```
